```python
import numpy as np 
import pandas as pd
import sys
from sklearn.tree import DecisionTreeClassifier
```

```python
my_data = pd.read_csv("training_3.csv", delimiter=",")

```

```python
#my_data= my_data[0:1000]
```

```python

#sys.getsizeof(my_data)/(1024*1024*1024)
#np.unique(my_data["PROV_B_M_A_STATE_NAME"])
```

```python
X = my_data[['ENTITY_TYPE_CODE',  'PROV_B_M_A_STATE_NAME', 'PROV_B_M_A_POSTAL_CODE', 'SPECDESC','PROVIDERNAME']].values
#X[0:3]
np.take(X,np.random.permutation(X.shape[0]),axis=0)
```

```python
from sklearn import preprocessing
le_STATE_NAME = preprocessing.LabelEncoder()
le_STATE_NAME.fit(np.unique(my_data["PROV_B_M_A_STATE_NAME"]))
X[:,1] = le_STATE_NAME.transform(X[:,1]) 

le_SPECDESC = preprocessing.LabelEncoder()
le_SPECDESC.fit(np.unique(my_data["SPECDESC"]))
X[:,3] = le_SPECDESC.transform(X[:,3]) 

le_PROVIDERNAME = preprocessing.LabelEncoder()
le_PROVIDERNAME.fit(np.unique(my_data["PROVIDERNAME"]))
X[:,4] = le_PROVIDERNAME.transform(X[:,4]) 


```

```python
#X[0:100]
```

```python
y = my_data["NPI"]
y[0:5]

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
X_trainset
```

```python
X_trainset.shape
```

```python
y_trainset.shape
```

```python
npiTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
npiTree # it shows the default parameters
```

```python
npiTree.fit(X_trainset,y_trainset)
```

```python
predTree = npiTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
